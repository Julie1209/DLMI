### [Matplotlib](https://matplotlib.org/)

````python
import numpy as np
import matplotlib.pyplot as plt
````

### plt.plot(): 折線圖

color: https://matplotlib.org/stable/gallery/color/named_colors.html

marker: https://matplotlib.org/stable/api/markers_api.html

````python
x = np.arange(0, 4*np.pi, 0.1) # [0, 0.1, 0.2 , ... 4pi]
y = np.sin(x)
print('x', x[:10])
print('y', y[:10])

plt.plot(x, y) # 繪製折線圖
plt.show() # 顯示圖表
````

````python
plt.plot(
    x, y,
    c='green', # 顏色
    marker="^", # 標記樣式
    alpha=0.3 # 透明度
)
plt.show()
````

### Basic

````python
x = np.arange(0, 4*np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.plot(x, y_sin) # 繪製第1條線
plt.plot(x, y_cos) # 繪製第2條線

# x, y軸標題
plt.xlabel("x axis label")
plt.ylabel("y axis label")
plt.title("Sine and Cosine") # 圖表標題
plt.legend(["Sine", "Cosine"]) # 圖例

plt.show()
````

### Subplot: 繪製子圖於figure中
```python
plt.subplot(row, column, 第幾i張圖)
```

NOTE: i從1開始

````python
plt.figure(figsize=(20, 3)) # 設定圖表大小 (寬, 高)
# 切換至 1x2 子圖中第1張圖
plt.subplot(1, 2, 1)
plt.plot(x, y_sin)
plt.title("Sine")
# 切換至 1x2 子圖中第2張圖
plt.subplot(1, 2, 2)
plt.plot(x, y_cos)
plt.title("Cosine")

plt.show()
````

### plt.scatter(): 散佈圖

````python
x = np.random.normal(0, 1, 1000)
y = np.random.normal(0, 3, 1000)
plt.scatter(x, y)
plt.title("Scatter plot")
plt.show()
````

````python
x = np.random.normal(0, 1, 1000)
y = np.random.normal(0, 1, 1000)
plt.scatter(x, y,
            s=100, # 標記大小
            c='blue', marker="v", alpha=0.5)
plt.title("Scatter plot")
plt.show()
````

# plt.bar(): 長條圖

````python
x = np.arange(10)
y = np.random.randint(0, 20, 10)
plt.bar(x, y)
plt.show()
````

### plt.hist(): Histogram(直方圖)

bins: 區間數量

````python
# 常態分佈
data = np.random.normal(loc=0, scale=1, size=1_000_000)
plt.hist(
    data,
    bins=100 # bins: 區間數量
)
plt.title("Histogram")
plt.xlabel("Value")
plt.ylabel("Counts")
plt.show()
````

````python
# 均勻分佈
data = np.random.uniform(size=1_000_000)
plt.hist(data, bins=30)
plt.title("Histogram")
plt.xlabel("Value")
plt.ylabel("Counts")
plt.show()
````

### 顯示圖片: plt.imshow

````python
# Black Image
img = np.random.randint(0, 256, (4, 4))
print(img)
plt.imshow(img)
plt.show()
````

color map: https://matplotlib.org/stable/tutorials/colors/colormaps.html

````python
# cmap: color map 色階
plt.imshow(img,
           cmap='gray')
plt.show()
````

````python
plt.imshow(img, cmap='coolwarm')
````

````python
# H, W, Channels(R,G,B)
img = np.random.randint(0, 256, (5, 5, 3))
print(img)
plt.imshow(img)
plt.show()
````
