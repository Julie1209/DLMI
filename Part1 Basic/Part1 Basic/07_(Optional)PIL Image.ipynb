# PIL (Python Imaging Library)

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 讀取圖片
image = Image.open('/content/test.jpg')
image_np = np.array(image) # to numpy
```

```python
type(image)
```

```python
plt.imshow(image)
plt.show()
```

```python
# PIL.Image.size (寬 高)
# np.array (高 寬 通道數)
image.size, image_np.shape
```

Image.mode: 圖片模式

RGB, L(灰階), RGBA, CMYK

```python
image.mode
```

```python
image_gray = image.convert('L')
image_gray.mode
```

```python
np.array(image_gray).shape
```

```python
plt.imshow(image_gray, cmap='gray')
plt.show()
```

調整大小: resize

```python
resize_image = image.resize((200, 200))
plt.imshow(resize_image)
plt.show()
```
