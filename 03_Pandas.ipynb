```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```
# Create a empty DataFrame
df = pd.DataFrame()
# add a column
df['name'] = ['A', 'B', 'C']
df
```

```
# add 2nd and 3rd column
df['height'] = [150, 180, 170]
df['weight'] = [45, 70, 64]
df
```

```
# Create DataFrame with Values
df = pd.DataFrame(
    {
        'name': ['A', 'B', 'C'],
        'height': [150, 180, 170],
        'weight': [45, 70, 64]
    }
)
df
```

```
# First 2 rows
df.head(2)
```

```
# Last 2 rows
df.tail(2)
```

```
# index and columns
print(df.index)
print(df.columns)
# RangeIndex(start=0, stop=3, step=1)
# Index(['name', 'height', 'weight'], dtype='object')
```

```
# convert to numpy array
df_numpy = df.to_numpy()
print(df_numpy.shape)
print(df_numpy)
# (3, 3)
# [['A' 150 45]
#  ['B' 180 70]
#  ['C' 170 64]]
```

```
df['height>160'] = df['height'] > 160
df
```
